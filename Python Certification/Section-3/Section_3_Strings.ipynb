**Operations on strings**

```python
#The + operator used against two or more strings produces a new string containing all the characters from its arguments
# + operator is not commutative
str1 = 'a'
str2 = 'b'
print(str1 + str2)
print(str2 + str1)
```

```python
''' * operator needs a string and a number as arguments;
 a new string created by the nth replication of the argument's string.'''
print(5 * 'a')
print('b' * 4)
```

```python
# example of shortcut operators += and *=
string1="ab"

string2="cd"
string1+=string2 #string1 will be overwritten by new string (concatenation of string1 and string2)
print(string1)

string2=string2*3
print(string2)
```

```python
# Demonstrating the ord() function.
#ord() -returns aspecific character's ASCII/UNICODE code point based on charcter
char_1 = 'a'
char_2 = ' '  # space
char_3="A"

print(ord(char_1))
print(ord(char_2))
print(ord(char_3))
```

```python
# chr() - The function takes a code point and returns its character.
# Demonstrating the chr() function.
print(chr(52)) #returns digit 4
print(chr(65))
print(chr(97))
print(chr(945))
```

```python
result1=chr(ord('A'))=="A"
result2=ord(chr(97))==97
print(result1)
print(result2)
```

```python
# Slices

alpha = "abdefg"

print(alpha[1:3])
print(alpha[3:])
print(alpha[:3])
print(alpha[::2])
print(alpha[1::2])
print("--------------\n")
print(alpha[-1::-1])
print(alpha[-1::])
print(alpha[5:0:-1])
print(alpha[3:-2])
print(alpha[-2:1:-1])
print(alpha[5:1:-2])

```

```python
# in and not in operators
alphabet = "abcdefghijklmnopqrstuvwxyz"
print("ghi" in alphabet)
print("xyz" not in alphabet)
print('Q' in alphabet)
print("lm" in alphabet)
print('g' not in alphabet)
```

```python
# index() method- searches the sequence from the beginning, in order to find the first element of the value specified in its argument.
str_index="My age is 42 and what is yours?"
print(str_index.index('4'))
print(str_index.index('y')) # returns index of first found charcter
print(str_index.index('s'))
print(str_index.index('is'))
print(str_index.index('age'))
#Note: the element searched for must occur in the sequence - its absence will cause a ValueError exception.
print(str_index.index('z')) ##generates a ValueError exception.
```

```python
#count() method-counts all occurrences of the element inside the sequence.
# Demonstrating the count() method:
print("abcabbcb".count("b"))
print('abcabc'.count("d"))
print(''.count('a')) #empty string doesnt generate any error
```

```python
#capitalize() method -if the first character inside the string is a letter
#(note: the first character is an element with an index equal to 0, not just the first visible character), it will be converted to upper-case;
#all remaining letters from the string will be converted to lower-case.

st="ABcdEf89zf"
print(st.capitalize())
```

```python
print('AbcD'.capitalize())
print('aBccdEFg'.capitalize())
print("aB56g&$@89U7J".capitalize())
print(" 123AB".capitalize())
print("".capitalize())
```
