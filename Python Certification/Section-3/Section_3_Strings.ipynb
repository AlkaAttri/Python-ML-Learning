```python
# Caesar cipher.
text = input("Enter your message: ")
cipher = ''
for char in text:
    if not char.isalpha():
        continue
    char = char.upper()
    code = ord(char) + 1
    if code > ord('Z'):
        code = ord('A')
    cipher += chr(code)

print(cipher)
```

```python
# Caesar cipher - decrypting a message.
cipher = input('Enter your cryptogram: ')
text = ''
for char in cipher:
    if not char.isalpha():
        continue
    char = char.upper()
    code = ord(char) - 1
    if code < ord('A'):
        code = ord('Z')
    text += chr(code)

print(text)
```

```python
# IBAN Validator.
#British: GB72 HBZU 7006 7212 1253 00
#French: FR76 30003 03620 00020216907 50
#German: DE02100100100152517108

iban = input("Enter IBAN, please: ")
iban = iban.replace(' ','')

if not iban.isalnum():
    print("You have entered invalid characters.")
elif len(iban) < 15:
    print("IBAN entered is too short.")
elif len(iban) > 31:
    print("IBAN entered is too long.")
else:
    iban = (iban[4:] + iban[0:4]).upper()
    iban2 = ''
    for ch in iban:
        if ch.isdigit():
            iban2 += ch
        else:
            iban2 += str(10 + ord(ch) - ord('A'))
    iban = int(iban2)
    if iban % 97 == 1:
        print("IBAN entered is valid.")
    else:
        print("IBAN entered is invalid.")
```

```python
#Program to encrypt the alphabets value by adding a number ranging from 1 to 25.
'''
Test data
Sample input:
abcxyzABCxyz 123
2
Sample output:
cdezabCDEzab 123

Sample input:
The die is cast
25
Sample output:
Sgd chd hr bzrs '''
line=input("enter one line text to encrypt")
shift=int(input("enter a shift value-a number from the range 1 to 25"))
line=list(line)
print(line)
text=""
for ch in line:
    if ch.isalpha() :
      ch_code = ord(ch) + shift # ch_code is an integer, representing the shifted ASCII value

      if ch.islower():
          if ch_code > ord('z'): # If shifted code goes beyond 'z'
              ch_code -= 26      # Wrap back by 26 positions (total alphabet length)
      elif ch.isupper():
          if ch_code > ord('Z'): # If shifted code goes beyond 'Z'
              ch_code -= 26      # Wrap back by 26 positions

      # Convert the final integer code back to a character and append
      text += chr(ch_code)
    else:
      # If the character is not an alphabet, append it as is
      text += ch

print(text)
```

```python
# the Digit of Life for the date.
#Digit of Life is a digit evaluated using somebody's birthday. It's simple - you just need to sum all the digits of the date. If the result contains more than one digit, you have to repeat the addition until you get exactly one digit. For example:
''' example
1 January 2017 = 2017 01 01=20170101
2 + 0 + 1 + 7 + 0 + 1 + 0 + 1 = 12
1 + 2 = 3  '''

birthday = input("Enter your birthday (digits only): ")

# Keep reducing until one digit remains
while len(birthday) > 1:
    total = 0
    for digit in birthday:
        total += int(digit)
    birthday = str(total)

print("Digit of Life:", birthday)

```
